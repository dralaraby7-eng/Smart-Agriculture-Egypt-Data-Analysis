```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
import streamlit as st

# Load data
fact = pd.read_csv("Fact.csv")

dim_Area = pd.read_csv("Dim_Area.csv") # Load Dim_Area.csv 1

dim_Year = pd.read_csv("Dim_Year.csv") # Load Dim_Year.csv 6
dim_Element = pd.read_csv("Dim_Element.csv") # Load Dim_Element.csv 3 
dim_Item = pd.read_csv("Dim_Item.csv") # Load Dim_Item.csv 5
dim_Domain = pd.read_csv("Dim_Domain.csv") # Load Dim_Domain.csv 2
dim_Flag = pd.read_csv("Dim_Flag.csv") # Load Dim_Flag.csv 4
dim_Soil = pd.read_csv("Soil.csv") # Load Dim_Soil.csv 8
dim_Parameter = pd.read_csv("Parameters.csv") # Load Dim_Parameter.csv 9
dim_Temp_Rain = pd.read_csv("Fact_Temp_Rain.csv") # Load Dim_Temp_Rain.csv 7

# Merge (Star Schema)
dim_area_full = dim_Area.merge(
    dim_Soil,
    left_on="Area",
    right_on="Country_Code",
    how="left"
)

#  ربط Year مع Rain و Parameters
dim_year_full = dim_Year \
    .merge(dim_Temp_Rain, left_on="Year Code", right_on="YEAR", how="left") \
    .merge(dim_Parameter, left_on="Year Code", right_on="Year", how="left")


df = fact.merge(dim_area_full, on="Area Code (M49)", how="left") \
         .merge(dim_year_full, on="Year Code", how="left") \
         .merge(dim_Element, on="Element Code", how="left") \
         .merge(dim_Item, on="Item Code", how="left") \
         .merge(dim_Domain, on="Domain Code", how="left") \
         .merge(dim_Flag, on="Flag", how="left")

# Preview
print(df.head())

df.to_csv("final.csv", index=False)
```

```python

Total_Production = df[df['Element Code'] == 5510]['Value_Correct'].sum()
Total_Area_Harvested = df[df['Element Code'] == 5312]['Value_Correct'].sum()
Total_Area_Irrigated = df[df['Element Code'] == 5110]['Value_Correct'].sum()
Non_Irrigated_Area = Total_Area_Harvested - Total_Area_Irrigated
Total_Crops = df['Item Code'].nunique()
AVG_Temperature = df['Average_Temp2M'].mean()
AVG_Rainfall = df['Average_Rain'].mean()
Yield = Total_Production / Total_Area_Harvested

print("Total Production:", Total_Production)
print("Total Area Harvested:", Total_Area_Harvested)
print("Total Area Irrigated:", Total_Area_Irrigated)
print("Non-Irrigated Area:", Non_Irrigated_Area)
print("Total Crops:", Total_Crops)
print("Average Temperature:", AVG_Temperature)
print("Average Rainfall:", AVG_Rainfall)
print("Yield:", Yield)
```
